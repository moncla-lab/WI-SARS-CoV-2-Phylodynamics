```python
import pandas as pd

# Read in the TSV file
df = pd.read_csv("greater_5k.tsv", sep="\t")

# Define a function to assign age bins
def assign_age_bin(age):
    if pd.isna(age):
        return None
    if str(age) == "0-17":
        return "0-17"
    try:
        age = float(age)
        if 18 <= age <= 25:
            return "18-25"
        elif 26 <= age <= 40:
            return "26-40"
        elif 41 <= age <= 65:
            return "41-65"
        elif age >= 66:
            return "66-100"
    except ValueError:
        return None

# Apply the binning function
df["age_bins"] = df["age"].apply(assign_age_bin)

# Save the updated DataFrame back to the TSV file
df.to_csv("greater_5k.tsv", sep="\t", index=False)

print("✅ Age bins assigned and file saved successfully.")
```

```python
import pandas as pd

# Read in the TSV file
df = pd.read_csv("greater_5k.tsv", sep="\t")

# Define a function to assign age bins
def assign_urban_rural_combined(Urban_or_Rural_USDA_Classification):
    if pd.isna(Urban_or_Rural_USDA_Classification):
        return None
    if str(Urban_or_Rural_USDA_Classification) == "Urban":
        return "urban-greater-MN"
    if str(Urban_or_Rural_USDA_Classification) == "Rural":
        return "rural-greater-MN"

# Apply the binning function
df["urban_rural_combined"] = df["Urban_or_Rural_USDA_Classification"].apply(assign_urban_rural_combined)

# Save the updated DataFrame back to the TSV file
df.to_csv("greater_5k.tsv", sep="\t", index=False)

print("✅ Classification assigned and file saved successfully.")
```
