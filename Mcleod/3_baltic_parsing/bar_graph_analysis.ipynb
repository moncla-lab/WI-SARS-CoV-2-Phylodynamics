```python
import matplotlib.pyplot as plt
import numpy as np

# --- Data from your builds ---
data = {
    "Titration #1 (1000_1500)": {
        "non-MN_mcleod": 423,
        "greater-MN_mcleod": 210,
        "mcleod_greater-MN": 118,
        "mcleod_non-MN": 58
    },
    "Titration #2 (1500_1500)": {
        "non-MN_mcleod": 291,
        "greater-MN_mcleod": 319,
        "mcleod_greater-MN": 144,
        "mcleod_non-MN": 86
    }
}

# --- Extract keys (transitions) and values ---
transitions = list(data["Titration #1 (1000_1500)"].keys())
build1_counts = [data["Titration #1 (1000_1500)"][t] for t in transitions]
build2_counts = [data["Titration #2 (1500_1500)"][t] for t in transitions]

# --- Set up positions ---
x = np.arange(len(transitions))  # label locations
width = 0.35  # width of the bars

# --- Plot ---
fig, ax = plt.subplots(figsize=(8,6))
bars1 = ax.bar(x - width/2, build1_counts, width, label="Titration #1 (1000_1500)", color="steelblue")
bars2 = ax.bar(x + width/2, build2_counts, width, label="Titration #2 (1500_1500)", color="orange")

# --- Labels & Formatting ---
ax.set_ylabel("Count of Strains")
ax.set_xlabel("Transition")
ax.set_title("Transitions based on Mcleod Status")
ax.set_xticks(x)
ax.set_xticklabels([t.replace("_", "  ") for t in transitions], rotation=20)
ax.legend()

# Add value labels above bars
def add_labels(bars):
    for bar in bars:
        height = bar.get_height()
        ax.annotate(f'{height}',
                    xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom', fontweight='bold')

add_labels(bars1)
add_labels(bars2)

plt.tight_layout()
plt.show()
```
